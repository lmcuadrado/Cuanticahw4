```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
N = 1000 #Numero de puntos
psi = np.zeros([N,2]) #Arreglo para guardar solución de psi
psi_inicial = np.array([0,1]) #Condiciones iniciales Impar
E = 0.0 #Inciaizo la variable global de energia
k = 100.0 #Constante de resorte
m = 1.0 #masa
w = (k/m)**0.5 #frecuencia
h = 1.0 #valor de hbar, por facilidad la puse como 1
L = 1.0 #longitud del potencial
X = 2.0*L #limite de rango a plotear
x = np.linspace(-X, X, N) #rango a plotear 
```

```python
def V(x):
    if abs(x)< L:
        return 0.5*k*x**2
    else:
        return 0.5*k*L**2

def modelo(psi, x):
    estado0 = psi[1]
    estado1 = (2.0*m/h**2)*(V(x) - E)*psi[0]
    return np.array([estado0, estado1])


def funcionDeOnda(energia):
    global psi
    global E
    E = energia
    psi = odeint(modelo, psi_inicial, x)
    return psi[-1,0]
```

```python
LimiteClasico=0.5*k*L**2

#Generamos un rango de energia
rangoE = np.linspace(0,k*L**2, 100)

psi_1 = []

for E1 in rangoE:
    psi_1.append(funcionDeOnda(E1))
```

```python
plt.plot(rangoE,psi_1, color="r")
plt.axhline(y=0, linestyle='-')
plt.axvline(x=LimiteClasico, linestyle='-')
plt.xlabel('Energia, $E$')
plt.ylabel('$\Psi(x = b)$')
```

```python
plt.plot(rangoE,psi_1)
plt.axhline(y=0, color='r', linestyle='-')
plt.axvline(x=LimiteClasico, color='r', linestyle='-')
plt.axis([0,100,-100,100])
plt.xlabel('Energia, $E$')
plt.ylabel('$\Psi(x = b)$')
#Zoom cerca al limite Clasico
```
