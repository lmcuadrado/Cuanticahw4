```python
import numpy as np
import matplotlib.pylab as plt

min_u = 0.0
max_u = 20.0
du=0.01
n= int((max_u - min_u)/du) 

#Creo los arreglos 
u = np.zeros(n)
psi = np.zeros(n)
d_psi = np.zeros(n)

#Defino los valores iniciales

eps=800
u[0] = min_u

#Pares
psi[0] = 1.0
d_psi[0] = 0.0

#a=dv/dt=-wx
def dd_psi(u,psi):          
    return ((u**2) - eps)*psi

#Solución analitica de la ecuación por el metodo de euler
for i in range(0,n-1):    
    u[i+1] = u[i] + du        
    d_psi[i+1] = d_psi[i] + dd_psi(u[i],psi[i])*du  
    psi[i+1] = psi[i] + du * d_psi[i] 
```

```python
plt.plot(u,psi)
plt.xlabel("u")
plt.ylabel("psi(u)")
plt.show()
```
