```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint , simps
from scipy.optimize import brentq
```

```python
k = 100.0 
m = 1.0
w = (k/m)**0.5   #frecuencia

h = 1.0 
E = 0.0          #Energía
L = 1.0          #Amplitud del potencial
cl= 0.5*k*L**2    #Límite clásico


x_max = 2.0*L 
N = 100 #Numero de puntos

#Arreglos
x = np.linspace(-x_max, x_max, N)  #Rango a graficar
psi = np.zeros([N,2])              #Arreglo para guardar solución de psi
psi_inicial = np.array([1,0])      #Valores iniciales para obtener una solución par
dE = np.linspace(0,k*L**2, 100)    #Rango de energia
```

```python
#Defino el portencial V
def V(x):
    if abs(x) < L:
        return 0.5*k*x**2
    else:
        return cl

#modelo una funcion que me de los valores de la primera y segunda derivada de psi
def model(psi, x):
    d_psi= psi[1]        #primera derivada
    dd_psi = (-(2.0*m)/(h**2))*(E - V(x))*psi[0] #segunda derivada
    return np.array([d_psi, dd_psi])

#Resuelvo la ecuación diferencial con las condiciones necesarias para que la solución sea par
def func(e):
    global E
    global psi
    E = e
    psi = odeint(model, psi_inicial, x)
    return psi[-1,0]

lista=[]
for i in dE:
    lista.append(func(i)) 
```

```python
plt.plot(dE,lista, color="g")
plt.axhline(y=0)
plt.axvline(x=cl)
plt.axis([0,100,-100,100])  #hago zoom sobre la gráfica
plt.xlabel('Energia, $E$')
plt.ylabel('$\Psi(x = b)$')
```

```python
#ANALISIS DE ENERGÍA PARA LA CONDICIÓN IMPAR

psi_inicial = np.array([0,1]) #Valores iniciales para obtener una solución impar

def func(e):
    global E
    global psi
    E = e
    psi = odeint(model, psi_inicial, x)
    return psi[-1,0]

lista=[]
for i in dE:
    lista.append(func(i)) 
```

```python
plt.plot(dE, lista, color="g")
plt.axhline(y=0)
plt.axvline(x=cl)
plt.axis([0,100,-100,100])
plt.xlabel('Energia, $E$')
plt.ylabel('$\Psi(x = b)$')
```

```python
#VALORES DE ENERGÍA PERMITIDOS 

def encontrarCeros(x,f):
    TodosCeros= []
    a = np.sign(f)
    for i in range(len(f)-1):
        if a[i]+a[i+1] == 0:
            cero = brentq(func, x[i], x[i+1])
            TodosCeros.append(cero)
    return TodosCeros

ceros = encontrarCeros(dE, lista)
```
